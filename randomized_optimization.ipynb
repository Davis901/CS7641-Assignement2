```python
import six
import sys
import time
import random
import networkx as nx
sys.modules['sklearn.externals.six'] = six
# import mlrose
import joblib
# sys.modules['sklearn.externals.joblib'] = j
import mlrose_hiive as mlrose
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
max_attempts= 1000
max_iterations= 10000
iterations = range(1, max_iterations+1
                   )
```

# Four Peaks Problem
Out of four algorithm, Genetic Algorithm stands out
But I am getting mimic as the highest

```python
fitness = mlrose.FourPeaks()

problem = mlrose.DiscreteOpt(length=100, fitness_fn=fitness, maximize=True)
```

### Randomized Hill Climbing

```python
print("RHC begins")
start = time.time()

rhc_best_state_1, rhc_best_fitness_1, rhc_fitness_values_1 = mlrose.random_hill_climb(problem, 
                                                            max_attempts = max_attempts, 
                                                            max_iters= max_iterations, 
                                                            curve=True, 
                                                            random_state=42,
                                                            restarts=100)

end = time.time()
rhc_time = end - start

rhc_fitness_values_1=[float(item[0]) for item in rhc_fitness_values_1]
print("Consumed time: ", rhc_time)
print("RHC best state: ", rhc_best_state_1)
print("RHC best fitness: ", rhc_best_fitness_1)
print(len(rhc_fitness_values_1))
```

### Simulated Annealing

```python
print("SA begins")
start = time.time()

sa_best_state_1, sa_best_fitness_1, sa_fitness_values_1 = mlrose.simulated_annealing(problem, 
                                                            max_attempts = max_attempts, 
                                                            max_iters=max_iterations, 
                                                            curve=True, 
                                                            random_state=42,
                                                            schedule=mlrose.GeomDecay(init_temp = 1, decay=0.1, min_temp=0.001))

end = time.time()
sa_time = end - start

sa_fitness_values_1=[float(item[0]) for item in sa_fitness_values_1]
print("Consumed time: ", sa_time)
print("SA best state: ", sa_best_state_1)
print("SA best fitness: ", sa_best_fitness_1)
```

### Genetic Algorithm

```python
print("GA begins")
start = time.time()

ga_best_state_1, ga_best_fitness_1, ga_fitness_values_1 = mlrose.genetic_alg(problem, 
                                                            max_attempts = max_attempts, 
                                                            max_iters=max_iterations, 
                                                            curve=True, 
                                                            random_state=42,
                                                            pop_size=100,
                                                            mutation_prob=0.5)

end = time.time()
ga_time = end - start

ga_fitness_values_1=[float(item[0]) for item in ga_fitness_values_1]
print("Consumed time: ", ga_time)
print("GA best state: ", ga_best_state_1)
print("GA best fitness: ", ga_best_fitness_1)

```

### mimic

```python
print("mimic begins")
start = time.time()

mm_best_state_1, mm_best_fitness_1, mm_fitness_values_1 = mlrose.mimic(problem, 
                                                            max_attempts = 500, 
                                                            max_iters=500, 
                                                            curve=True, 
                                                            random_state=42,
                                                            keep_pct=0.1)

end = time.time()
mm_time = end - start

mm_fitness_values_1=[float(item[0]) for item in mm_fitness_values_1]
print("Consumed time: ", mm_time)
print("MM best state: ", mm_best_state_1)
print("MM best fitness: ", mm_best_fitness_1)

```

# Max K color problem

stands out Mimic algorithm

```python
# edges = [
#     # Cluster 1: Complete connections
#     (0, 1), (0, 2), (1, 2),
#     # Cluster 2: A chain
#     (3, 4), (4, 5), (5, 6), (6, 7),
#     # Cluster 3: Star structure
#     (8, 9), (8, 10), (8, 11),
#     # Inter-cluster connections: Create dependencies across clusters
#     (2, 3), (7, 8),
#     # Additional cross-cluster edges to increase complexity
#     (1, 9), (0, 10), (2, 11), (6, 9), (5, 10)
# ]

G = nx.Graph()
    
cluster_sizes = [10, 10, 10]
base = 0
for size in cluster_sizes:
    cluster = nx.connected_watts_strogatz_graph(n=size, k=4, p=0.5, tries=100)
    cluster = nx.relabel_nodes(cluster, {i: i + base for i in range(size)})
    base += size
    G = nx.compose(G, cluster)
inter_cluster_edges = [(5, 15), (5, 25), (15, 20), (20, 25)]
G.add_edges_from(inter_cluster_edges)
edges = list(G.edges)

print(f"Generated {len(edges)} edges")
# Define the fitness function
fitness_2 = mlrose.MaxKColor(edges=edges)

# Define the optimization problem object
problem_2 = mlrose.DiscreteOpt(length=30,  # Number of nodes
                             fitness_fn=fitness_2,
                             maximize=True)
                            #  max_val=3)  # Max number of colors
```
